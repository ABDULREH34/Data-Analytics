```python

import matplotlib.pyplot as plt

# Checking Sum of Units Sold Data
sum_check = df['Item Type'].value_counts()

# Set the size of the plot
plt.figure(figsize=(12, 8))

# Plot the pie chart
plt.pie(
    sum_check,
    labels=sum_check.index,
    autopct='%1.1f%%',
    textprops={'fontsize': 10},
    wedgeprops={'linewidth': 0.5}
)

# Ensure the pie chart is a circle
plt.axis('equal')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Top 10 countries
top_10_country = df['Country'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 countries
filtered_df = df[df['Country'].isin(top_10_country)]

# Plot count of countries
plt.figure(figsize=(12, 6))
sns.countplot(data=filtered_df, x='Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Distribution of Top 10 Countries')
plt.xticks(rotation=45)
plt.show()
```
